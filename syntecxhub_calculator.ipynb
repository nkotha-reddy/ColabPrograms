```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        print("Cannot divide by zero")
        return None
    return a / b

def calculate(a, operator, b):
    if operator == '+':
        return add(a, b)
    elif operator == '-':
        return subtract(a, b)
    elif operator in ['*', 'x', 'X']:
        return multiply(a, b)
    elif operator in ['/', '÷']:
        return divide(a, b)
    else:
        print("Invalid operator")
        return None

def parse_input(user_input):
    parts = user_input.strip().split()

    if len(parts) != 3:
        print("Input must be: number operator number")
        return None

    try:
        num1 = float(parts[0])
        operator = parts[1]
        num2 = float(parts[2])
        return num1, operator, num2
    except:
        print("Invalid number format")
        return None

def show_menu():
    print("\nCommand Line Calculator")
    print("Enter expression (e.g. 2 + 3)")
    print("Operators: + - * /")
    print("Type 'clear' to reset")
    print("Type 'exit' to quit")

def main():
    while True:
        show_menu()
        user_input = input(">> ").strip().lower()

        if user_input == 'exit':
            print("Goodbye!")
            break
        if user_input == 'clear':
            print("Calculator cleared")
            continue

        parsed = parse_input(user_input)
        if not parsed:
            continue

        num1, operator, num2 = parsed
        result = calculate(num1, operator, num2)

        if result is not None:
            print(f"Result: {result}")

if __name__ == "__main__":
    main()
```
